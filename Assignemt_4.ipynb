### 1. Maximum Product of Two Elements in an Array

```python
def maxProduct(nums):
    output = 0

    for i in range(len(nums)):
        for j in range(i+1, len(nums)):
            output = max(output, (nums[i] - 1) * (nums[j] - 1))

    return output

nums = [1,5,4,5]
maxProduct(nums)
```

### 2. Count Number of Teams

```python
def numTeam(rating):
    n = len(rating)
    teams = 0

    for j in range(1, n-1):
        left_smaller = left_greater = 0
        right_smaller = right_greater = 0

        for i in range(j):
            if rating[i] < rating[j]:
                left_smaller += 1
            elif rating[i] > rating[j]:
                left_greater += 1

        for k in range(j + 1, n):
            if rating[k] > rating[j]:
                right_greater += 1
            elif rating[k] < rating[j]:
                right_smaller += 1

        teams += left_smaller * right_greater
        teams += left_greater * right_smaller

    return teams

rating = [2, 5, 3, 4, 1]
numTeam(rating)
```

### 3. Number of Students Doing Homework at a Given Time

```python
def busyStudent(startTime, endTime, queryTime):
    count = 0

    for i in range(len(startTime)):
        if startTime[i] <= queryTime <= endTime[i]:
            count += 1

    return count

startTime = [1,2,3]
endTime = [3,2,7]
queryTime = 4
busyStudent(startTime, endTime, queryTime)
```

### 4. Number of Steps to Reduce a Number to Zero

```python
def numOfSteps(num):
    step = 0

    while num > 0:
        if num % 2 == 0:
            num //= 2
        else:
            num -= 1
        step += 1

    return step
    
num = 14
numOfSteps(num)
```

### 5. Counting Bits

```python
def countBits(n):
    ans = [0] * (n + 1)

    for i in range(1, n + 1):
        ans[i] = ans[i // 2] + (i % 2)

    return ans

n = int(input())
countBits(n)
```
